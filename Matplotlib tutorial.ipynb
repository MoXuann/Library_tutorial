```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
x = np.linspace(0,5,11)
y = x ** 2
```

FUNCTIONAL

```python
##FUNCTIONAL
plt.plot(x,y,'r-')
plt.xlabel("X label")
plt.ylabel('Y label')
plt.title("title")
```

```python
#subplot
plt.subplot(1,2,1)
plt.plot(x,y,"r-")

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

OBJECT-ORIENTED

```python
#OO
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
```

Draw two axes in one plot

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

Subplots

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

# for current_ax in axes:
#     current_ax.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title('First')

axes[1].plot(y,x)
axes[1].set_title('Second')
```

Change the figsize

```python
fig =  plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

plt.tight_layout()
```

Plot two lines in one graph

```python
fig =  plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='X Square')
ax.plot(x, x**3, label='X Cube')

ax.legend()
```

Save the matplot figure

```python
fig.savefig('hihi.png',dpi=200)
```

```python
fig =  plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='#FF8C00', lw=2, alpha=0.5, ls=":", marker="o", markersize=2, markerfacecolor="yellow") #RGB Hex Code or color name
#lw == linewidth
#ls == linestyle

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
